```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from tqdm import tqdm

sns.set()
sns.set_context("poster")
```

```python
def f(x, r):
    return r * x * (1-x)

def logistic_map(x, r, order=1):
    y = x
    for i in range(order):
        y = f(y, r)
    return y
    

x = np.linspace(-0.1, 1.1, 100)

# First Order Mapping
plt.figure(figsize=(10, 10))
plt.plot(x, x, "--r")
plt.plot(x, logistic_map(x, 1))
plt.plot(x, logistic_map(x, 2))
plt.plot(x, logistic_map(x, 3))
plt.plot(x, logistic_map(x, 4))
plt.title("First Order Mapping: $y = f(x)$")
plt.legend(["$y=x$", "r=1", "r=2", "r=3", "r=4"])
plt.ylim([-0.2, 1.1])

# Second Order Mapping
plt.figure(figsize=(10, 10))
plt.plot(x, x, "--r")
plt.plot(x, logistic_map(x, 1, order=2))
plt.plot(x, logistic_map(x, 2, order=2))
plt.plot(x, logistic_map(x, 3, order=2))
plt.plot(x, logistic_map(x, 4, order=2))
plt.title("Second Order Mapping: $y = f(x)$")
plt.legend(["$y=x$", "r=1", "r=2", "r=3", "r=4"])
plt.ylim([-0.2, 1.1])
```

# Bifurcation Diagram 

```python
N_POINTS = 1000
N_ITERATION = 1000

r = np.linspace(1, 4, 10000)
points = np.zeros((len(r), N_POINTS))

for i, r_now in enumerate(tqdm(r)):
    points_now = np.random.rand(N_POINTS)
    for _ in range(N_ITERATION):
        points_now = logistic_map(points_now, r_now)
    points[i] = points_now

plt.figure(figsize=(20, 10))
for i, r_now in enumerate(r):
    plt.scatter(x=[r_now] * N_POINTS, y=points[i], s=0.1, c="b", alpha=0.5)
```
